```python
import csv

def csv_to_dictionary(filename):
    data = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(row)
    return data

csv_filename = '/kaggle/input/iphone-features/f2.csv'
iphone_data = csv_to_dictionary(csv_filename)

# Accessing information for a specific iPhone model (case-insensitive)
iphone_model = input('iPhone Model:')
matching_models = [item for item in iphone_data if item["Model"].lower() == iphone_model.lower()]

if matching_models:
    iphone_info = matching_models[0]
    print(iphone_info)
else:
    print(f"No matching iPhone model found for '{iphone_model}'")
```
